**Problem Introduction:** This analysis investigates whether STA130 students can correctly identify whether milk or tea is poured first to replicate the experiment by Ronald Fisher and Muriel Bristol. The data collected shows that 49 out of 80 students were able to guess correctly.

The test with the STA130 students will provide a much more standardized result compared to Dr. Fishers test due to the larger sample size and the lack of anomolies within the sample such as Dr. Bristol who may have a greater acuity to guessing the order correctly compared to the average STA130 student.

Null Hypothesis:

Formal Version: The proportion of students who can guess the order correctly will be 0.5. This is stated mathematically as $H_0$ : p = 0.5

Informal Version: The STA130 students will simply guess with no added reasoning which will lead to around 50% guessing correctly.

Alternative Hypothesis:

The proportion of students who can guess the order correctly will not be 0.5. This is stated mathematically as $H_A$ $\not =$ $H_0$


**Quantitative Analysis**

```python
# Import necessary libraries
import numpy as np
from scipy import stats

# Problem Introduction
# STA130 students are tested to see if they can distinguish whether tea or milk was poured first. 
# 49 out of 80 students correctly identified the correct pouring order.

# Given Data
n = 80  # Sample size
correct_identifications = 49  # Number of correct identifications

# Sample proportion (observed test statistic)
p_hat = correct_identifications / n  # Sample proportion of correct identifications
print(f"Sample proportion (pÃÇ): {p_hat:.4f}")

# Null Hypothesis (H‚ÇÄ): p = 0.5 (random guessing)
# Alternative Hypothesis (H‚ÇÅ): p > 0.5 (students identify better than random guessing)

p_0 = 0.5  # Proportion under the null hypothesis

# Calculate z-score
z_score = (p_hat - p_0) / np.sqrt(p_0 * (1 - p_0) / n)
print(f"Z-score: {z_score:.4f}")

# Calculate the p-value (one-tailed test for p > 0.5)
p_value = 1 - stats.norm.cdf(z_score)
print(f"P-value: {p_value:.4f}")

# Interpretation of results
alpha = 0.05  # Significance level

if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("There is evidence that students identify whether tea or milk was poured first better than by random guessing.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("There is insufficient evidence to conclude that students identify the pouring order better than random guessing.")

# Optional: Supporting Visualizations (not necessary but could be helpful)
# We can visualize a normal distribution with the z-score

import matplotlib.pyplot as plt
import seaborn as sns

# Generate data for the normal distribution
x = np.linspace(-3, 3, 500)
y = stats.norm.pdf(x)

# Plot the standard normal distribution
plt.figure(figsize=(8, 5))
sns.lineplot(x=x, y=y, color="blue")

# Shade the region under the curve from z-score to infinity (one-tailed test)
plt.fill_between(x, 0, y, where=(x >= z_score), color="red", alpha=0.3)

# Plot labels and title
plt.title(f"Z-Test for Proportions: Z-Score = {z_score:.2f}, P-value = {p_value:.4f}")
plt.xlabel("Z-Score")
plt.ylabel("Density")

# Display plot
plt.show()

# Make the analysis reproducible (optional but recommended)
np.random.seed(42)
```
